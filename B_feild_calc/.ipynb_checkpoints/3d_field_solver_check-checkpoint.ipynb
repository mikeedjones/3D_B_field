```python
B2d=s2d.print_fields_2d([s2d.solenoid_2d_map(2,0,0,0,1,1,r2d,1)[0][2]],r2d,[[2,2,1]])
plt.clf()
r3d=s3d.bio.r_grid([0,1],[0,1],[-0.5,0.5],n)

std_diff=[]
std_div=[]
mean_diff=[]
mean_div=[]

set_N=np.arange(10,101,3)

for N in set_N:
    h=s3d.solenoid(2,0,0,0,1,1,N,0)
    hc=s3d.polar_2_cart(h)
    ht=hc.transpose()
    B3d=s3d.bio.trapping_field(hc,r3d,n=N)
    Bslicex, Bslicey, Bslicez=s3d.bio.print_fields(*B3d,[222,222,222])
    B2d_mag=B2d[0]
    diffy=np.nan_to_num((B2d_mag-Bslicey)/Bslicey)
    divy=B2d_mag/Bslicey
    
    mean_div.append(np.mean(divy))
    mean_diff.append(np.mean(diffy[np.where(diffy>-1)]))
    
    std_div.append(np.std(divy))
    std_diff.append(np.std(diffy[np.where(diffy>-1)]))

plt.clf()

plt.errorbar(set_N,mean_div,yerr=std_div,marker='.',linestyle='None')
plt.errorbar(set_N,mean_diff,yerr=std_diff,marker='.',linestyle='None')
```
